```python
def temps(time_1, time_2):   
    sec1 = 3600*int(time_1[0:2]) + 60*int(time_1[3:5]) + int(time_1[6:8])
    sec2 = 3600*int(time_2[0:2]) + 60*int(time_2[3:5]) + int(time_2[6:8])                                                                                                                                   
    return (sec2-sec1)

def distance(x1,x2,y1,y2):
    dist = np.sqrt(((x2-x1)**2)+((y2-y1)**2))
    return dist

def velocity(dist, time):
    if time == 0:
        return 0
    else:
        v = dist / time
        return v
```

```python
def conv(time_1):   
    sec1 = 3600*int(time_1[0:2]) + 60*int(time_1[3:5]) + int(time_1[6:8])                                                                                                                                
    return (sec1)
```

```python
vec_conv =  np.vectorize(conv)
```
